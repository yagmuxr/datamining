```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parameters for the beta distribution
a_values = [0.5, 2, 5]  # Different 'a' parameters
b_values = [0.5, 2, 5]  # Corresponding 'b' parameters

# Create a range of x values between 0 and 1
x = np.linspace(0, 1, 100)

# Plotting the PDF
plt.figure(figsize=(12, 6))

for a, b in zip(a_values, b_values):
    plt.plot(x, beta.pdf(x, a, b), label=f'Beta(a={a}, b={b})')

plt.title('Beta Distribution PDF for Various Parameters')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()

# Sampling from the beta distribution
samples = np.random.beta(2, 5, size=1000)

# Histogram of sampled data
plt.figure(figsize=(12, 6))
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g', label='Sampled Data')

# Overlay the theoretical PDF
plt.plot(x, beta.pdf(x, 2, 5), 'r-', label='Theoretical PDF (a=2, b=5)')
plt.title('Histogram of Samples from Beta Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

```python
Random Types
In the context of random number generation, there are various types of distributions from which you can generate random numbers. 
Here are a few common types:

Uniform Distribution: All values in a specified range are equally likely.

Example: np.random.uniform(low, high, size)
Normal (Gaussian) Distribution: Characterized by its bell-shaped curve, defined by a mean and standard deviation.

Example: np.random.normal(loc, scale, size)
Binomial Distribution: Represents the number of successes in a fixed number of independent Bernoulli trials.

Example: np.random.binomial(n, p, size)
Poisson Distribution: Models the number of events in a fixed interval of time or space, given a known average rate of occurrence.

Example: np.random.poisson(lam, size)
Exponential Distribution: Describes the time between events in a Poisson process, characterized by a rate parameter.

Example: np.random.exponential(scale, size)
Beta Distribution: As mentioned earlier, models random variables constrained to [0, 1].

Beta Distribution
The beta distribution is a continuous probability distribution defined on the interval [0, 1]. 
It is characterized by two shape parameters, often denoted as 'a' and 'b'. 
This distribution is widely used in statistics, particularly in Bayesian analysis, to model random variables that are constrained to the interval [0, 1]

Beta Distribution Overview
The probability density function (PDF) of the beta distribution is given by:
f(x; a, b) = (x^(a-1) * (1-x)^(b-1)) / B(a, b)
where:
- x is a random variable in the interval [0, 1].
- B(a, b) is the beta function, which normalizes the distribution.
Parameters a and b
1. Parameter a:
- Controls the behavior of the distribution near the left end of the interval (0).
- Interpreted as the number of 'successes' in a Bayesian context or as a concentration parameter.
- If a > 1, the distribution is more likely to produce values closer to 1.
- If a < 1, the distribution favors values closer to 0.
- If a = 1, the distribution is uniform across the interval [0, 1].

2. Parameter b:
- Controls the behavior of the distribution near the right end of the interval (1).
- Similar to a, b can be viewed as the number of 'failures' in a Bayesian context.
- If b > 1, the distribution is more likely to produce values closer to 0.
- If b < 1, the distribution favors values closer to 1.
- If b = 1, the distribution is uniform.
Visualizing Different Combinations
1. Both a and b greater than 1:
- The distribution is bell-shaped and peaked in the middle of the interval.
- Example: a=5, b=5 creates a distribution that peaks around 0.5.

2. Both a and b less than 1:
- The distribution is U-shaped, with higher probabilities at the extremes (0 and 1).
- Example: a=0.5, b=0.5 creates a distribution where values close to 0 and 1 are more likely.

3. One parameter greater than 1 and the other less than 1:
- The distribution is skewed toward one side.
- Example: a=2, b=5 creates a distribution skewed toward 0, while a=5, b=2 skews toward 1.

4. Both a and b equal to 1:
- The distribution is uniform across the interval [0, 1].
```
