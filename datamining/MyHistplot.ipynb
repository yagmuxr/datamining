```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df_animal = pd.read_csv('data/msleep.csv')

def my_histplot(data, bins=8,title='title',xlabel='values',ylabel='counts'):
    # Min ve max değerleri bul
    data_min = min(data)
    data_max = max(data)

    # Aralıkları (bins) oluştur
    bin_width = (data_max - data_min) / bins
    bin_edges = np.linspace(data_min, data_max, bins + 1)  # Aralık sınırlarını oluştur

    # Her bin içerisindeki veri sayısını hesapla
    bin_counts = np.zeros(bins)
    for value in data:
        # Hangi aralığa düştüğünü bul
        bin_index = int((value - data_min) / bin_width)
        if bin_index == bins:  # max değerin tam olarak son bin'e düşmesi durumu
            bin_index = bins - 1
        bin_counts[bin_index] += 1

    # Histogramı çizdir
    plt.bar(bin_edges[:-1], bin_counts, width=bin_width, edgecolor='black', align='edge')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()

my_histplot(data=df_animal.sleep_total,title='Distribution of Sleep Times of Various Mammals',xlabel='Hours of Sleep')
```
