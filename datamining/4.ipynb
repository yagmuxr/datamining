```python
#binomial dağılım
#bir dizi bağımsız denemedeki başarı sayısının olasığını tanımlar. 
# Ayrık bir dağılımdır sayılabilir.
# n deney sayısı p başarı olasılığı expected value = n *p
# Bağımsız olay olmalı ve 2 seçenek olmalı

from scipy.stats import binom

binom.rvs(8, 0.5, size = 5)
```

```python
binom.rvs(3, 0.5, size = 10) # elimizde 3 tane para var aynı anda 10 kez havaya atıyoruz kaç başarılı sonuç elde ederiz
```

```python
#normal dağılım
# iki parametresi vardır sapma ve mü değeri
# uçlardaki ihtimal hiçbir zaman 0 olmaz 
# verilerin %68 ortalamanın 1 std sapma aşağısı ve yukarısı verir
# %95 2 alt 2 üst
# %99.7 3 alt 3 üst yapar
# outlier hesaplarken ise M-3std<veri M+3std>veri olarak da hesaplanabilir.
# 68 - 95 - 99.7 kuralın sıraylar ortalamadan 1,2,3 std uzakta demektir.
# ortalama 0 sapması 1 olan normal dağılımı standart normal dağılım denir.
from scipy.stats import norm

norm.cdf(154,161,7) # ortalaması ve sapması sırayısla 161 ve 7 olan bir normal dağılımda 154 ve altında boyu olan bir kadınların olasılığı
```

```python
1 - norm.cdf(154,161,7)
```

```python
norm.cdf(157,161,7) - norm.cdf(154,161,7)
```

```python
# verilen bir olasılığa karşılık gelen değeri bulur
norm.ppf(0.9,161,7) # kadınları %90 ı 169.97 den kısadır
```

```python
norm.rvs(161,7, size=10) #161 ort 7 sapmaya ait bir normal dağılımdan 10 tane rastgele değer üretme
```

```python
#kadınların %90 ı 152 boyundan uzundur
norm.ppf(0.1,161,7)
```

```python
#merkezi limit teoremi
# ortalama gibi bir özet istatistiğinin dağılımına örneklem dağılımı denir.
# deneme sayısı arttıkça grafik normal dağılıma yaklaşır(merkezi limit teoremi)
# standart sapma ve oranda da örneklem büyüklüğü arttıkça merkezi limit teoremi işler.
# örneklem büyüklüğü en az 30 olmalı

import pandas as pd
import numpy as np

die = pd.Series([1,2,3,4,5,6])
samp_5 = die.sample(5,replace=True)
samp_5
```

```python
import matplotlib.pyplot as plt

sample_means = []

for i in range(10):
    samp_5 = die.sample(5,replace=True)
    sample_means.append(samp_5.mean())
plt.title("Örneklem Ortalamasının Örneklem Dağılımı")
plt.hist(sample_means)
```

```python
import pandas as pd 
sales_team = pd.Series(['Amir','Brian','Claire','Damian'])
sales_team.sample(10,replace=True)
```

```python
sample_prp = []
for i in range(1000):
    samp_5 = sales_team.sample(10,replace=True)
    try:
        sample_prp.append(samp_5.value_counts()['claire']/10)
    except:
        sample_prp.append(0)
plt.title("Örneklem Oran Örneklem dağılımı")
plt.hist(sample_prp)
```

```python
# poisson_distribution
# ayrık dağılımdır
from scipy.stats import poisson

#Haftada ortalama 8 sahiplenmenin gerçekleştirildiği bir barınakta bir haftada 5 sahiplendirme gerçekleşme olasılığı nedir
poisson.pmf(5,8)

# 5 veya daha az sahiplenme olasılığı
poisson.cdf(5,8)

#5 den daha fazla sahiplenme olasılığı
1 - poisson.cdf(5,8)
```

```python
# rastgele değer üretmeyi sağlıyor
# lambda 8 miktar 10
poisson_arr = poisson.rvs(8, size=10)
poisson_arr
```

```python
# üstel dağılım
# sürekli dağılımdır
# poisson olaylarında belli bir zaman geçme olasılığı

# depremler arasında 6-8 ay geçme olasılığı
# lambda dağılımın ne kadar dik bir şekilde dağılacağını etkiliyor
from scipy.stats import expon

# yeni bir istek için 1 dakikadan az bekleme olasılığı
expon.cdf(1,scale=2)

# 4 dakikadan fazla bekleme olasılığı
expon.cdf(4,scale=2)

# 1 ile 4 dakika arasında bekleme olasılığı
expon.cdf(4,scale=2) - expon.cdf(1,scale=2)
```

```python
# t dağılımı
# küçük orneklem boyutları ile uğraşırken popülasyonun sapması bilinmediğinde yaygın kullanılır


#serbestlik derecesi bir popülasyonun bir parametresini kaç tane bilgi ile tahmin edebiliriz. 
# düşük serbestlik derecesinde veriler sağa sola yayılır
# yüksek sd dağılım normal dağılıma yaklaşır
# sd n-1 olarak alınır n örneklem boyutu

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

x = np.linspace(-5, 5, 100)

# Plot for different degrees of freedom
for df in [1, 5, 10, 30]:
    plt.plot(x, stats.t.pdf(x, df), label=f'df={df}')

plt.plot(x, stats.norm.pdf(x), label='Normal distribution', 
         linestyle='--')
plt.legend()
plt.title('t-distribution with different degrees of freedom')
plt.xlabel('x')
plt.ylabel('Density')
plt.show()
```
