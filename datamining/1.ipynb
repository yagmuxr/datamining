```python
pandas.DataFrame.sort_values() fonksiyonu, bir DataFrame'i belirli bir sÃ¼tuna (veya sÃ¼tunlara) gÃ¶re sÄ±ralamak iÃ§in kullanÄ±lÄ±r. 
Bu fonksiyonun parametreleri, sÄ±ralama iÅŸlemi Ã¼zerinde oldukÃ§a fazla esneklik saÄŸlar. Ä°ÅŸte bu parametrelerin aÃ§Ä±klamasÄ± ve Ã¶rnekleri:

Parametreler
by: (string veya list)

SÄ±ralamak istediÄŸiniz sÃ¼tunun adÄ±nÄ± veya sÃ¼tunlarÄ±n adlarÄ±ndan oluÅŸan bir listeyi belirtir.
Ã–rnek:
df.sort_values(by="age")  # "age" sÃ¼tununa gÃ¶re sÄ±ralar
df.sort_values(by=["age", "salary"])  # Ã–nce "age", sonra "salary" sÃ¼tununa gÃ¶re sÄ±ralar
axis: (0 veya 1, default = 0)

SÄ±ralamanÄ±n satÄ±rlar (0) veya sÃ¼tunlar (1) Ã¼zerinde yapÄ±lacaÄŸÄ±nÄ± belirler. Genellikle satÄ±rlar sÄ±ralanÄ±r.
Ã–rnek:
df.sort_values(by="age", axis=0)  # SatÄ±rlarÄ± "age" sÃ¼tununa gÃ¶re sÄ±ralar (standart kullanÄ±m)
df.sort_values(by="age", axis=1)  # SÃ¼tunlarÄ± "age" deÄŸerine gÃ¶re sÄ±ralar
ascending: (bool veya list, default = True)

SÄ±ralamanÄ±n artan (True) veya azalan (False) olacaÄŸÄ±nÄ± belirtir. Birden fazla sÃ¼tun sÄ±ralamasÄ± iÃ§in liste de verebilirsiniz.
Ã–rnek:
df.sort_values(by="age", ascending=True)  # "age" sÃ¼tununu artan sÄ±raya gÃ¶re sÄ±ralar
df.sort_values(by="age", ascending=False)  # "age" sÃ¼tununu azalan sÄ±raya gÃ¶re sÄ±ralar
df.sort_values(by=["age", "salary"], ascending=[True, False])  # "age" artan, "salary" azalan sÄ±ralanÄ±r
inplace: (bool, default = False)

True yaparsanÄ±z, orijinal DataFrame'i yerinde sÄ±ralar (yeni bir DataFrame dÃ¶ndÃ¼rmez, mevcut olanÄ± deÄŸiÅŸtirir).
Ã–rnek:
df.sort_values(by="age", inplace=True)  # DataFrame'i kalÄ±cÄ± olarak sÄ±ralar
kind: (string, default = 'quicksort')

KullanÄ±lacak sÄ±ralama algoritmasÄ±nÄ± belirtir. SeÃ§enekler:
'quicksort': HÄ±zlÄ±dÄ±r, en yaygÄ±n kullanÄ±lan.
'mergesort': KararlÄ± bir sÄ±ralama yapar (aynÄ± deÄŸerlere sahip olanlar orijinal sÄ±rayÄ± korur).
'heapsort': Daha Ã§ok teorik avantajlar saÄŸlar, pratikte daha az kullanÄ±lÄ±r.
Ã–rnek:
df.sort_values(by="age", kind="mergesort")  # "age" sÃ¼tununu kararlÄ± bir ÅŸekilde sÄ±ralar
na_position: ('first' veya 'last', default = 'last')

NaN (eksik deÄŸerler) sÄ±ralanÄ±rken nereye konulacaÄŸÄ±nÄ± belirtir.
'first': NaN deÄŸerler en baÅŸa konur.
'last': NaN deÄŸerler en sona konur (varsayÄ±lan).
Ã–rnek:
df.sort_values(by="age", na_position="first")  # NaN deÄŸerler sÄ±ralamanÄ±n baÅŸÄ±nda yer alÄ±r
ignore_index: (bool, default = False)

SÄ±ralandÄ±ktan sonra orijinal indekslerin korunup korunmayacaÄŸÄ±nÄ± belirtir. True yaparsanÄ±z, sÄ±ralamadan sonra indeksler sÄ±fÄ±rlanÄ±r.
Ã–rnek:
df.sort_values(by="age", ignore_index=True)  # SÄ±ralamadan sonra indeksler sÄ±fÄ±rlanÄ±r
key: (fonksiyon, default = None)

Her sÃ¼tundaki deÄŸerlere sÄ±ralamadan Ã¶nce bir fonksiyon uygulamak iÃ§in kullanÄ±lÄ±r. 
Bu parametre Python'Ä±n sorted() fonksiyonundaki key parametresine benzer.
Ã–rnek:
df.sort_values(by="age", key=lambda x: x.abs())  # Negatif deÄŸerleri pozitif deÄŸerlere dÃ¶nÃ¼ÅŸtÃ¼rerek sÄ±ralar
Ã–rneklerle AÃ§Ä±klamalar
Ã–rnek 1: Temel KullanÄ±m

import pandas as pd

data = {'name': ['Ali', 'Veli', 'AyÅŸe', 'Fatma'],
        'age': [25, 32, 19, 29]}

df = pd.DataFrame(data)

# YaÅŸ sÃ¼tununa gÃ¶re sÄ±ralama
df_sorted = df.sort_values(by='age')
print(df_sorted)
Ã‡Ä±ktÄ±:

    name  age
2  AyÅŸe   19
0  Ali    25
3  Fatma  29
1  Veli   32
Ã–rnek 2: Azalan SÄ±ralama
python
Kodu kopyala
df_sorted_desc = df.sort_values(by='age', ascending=False)
print(df_sorted_desc)
Ã‡Ä±ktÄ±:

    name  age
1  Veli   32
3  Fatma  29
0  Ali    25
2  AyÅŸe   19
Ã–rnek 3: Birden Fazla SÃ¼tun ile SÄ±ralama
python
Kodu kopyala
data = {'name': ['Ali', 'Veli', 'AyÅŸe', 'Fatma'],
        'age': [25, 32, 19, 25],
        'salary': [5000, 6000, 2000, 5000]}

df = pd.DataFrame(data)

# YaÅŸ ve maaÅŸa gÃ¶re sÄ±ralama (Ã¶nce yaÅŸ, yaÅŸlar eÅŸit olursa maaÅŸ)
df_sorted = df.sort_values(by=['age', 'salary'], ascending=[True, False])
print(df_sorted)
Ã‡Ä±ktÄ±:

    name  age  salary
2  AyÅŸe   19    2000
0  Ali    25    5000
3  Fatma  25    5000
1  Veli   32    6000
Bu parametrelerle sort_values() fonksiyonunu esnek ve gÃ¼Ã§lÃ¼ bir ÅŸekilde kullanabilirsiniz!
```
