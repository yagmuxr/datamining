```python
pandas.DataFrame.sort_values() fonksiyonu, bir DataFrame'i belirli bir sütuna (veya sütunlara) göre sıralamak için kullanılır. 
Bu fonksiyonun parametreleri, sıralama işlemi üzerinde oldukça fazla esneklik sağlar. İşte bu parametrelerin açıklaması ve örnekleri:

Parametreler
by: (string veya list)

Sıralamak istediğiniz sütunun adını veya sütunların adlarından oluşan bir listeyi belirtir.
Örnek:
df.sort_values(by="age")  # "age" sütununa göre sıralar
df.sort_values(by=["age", "salary"])  # Önce "age", sonra "salary" sütununa göre sıralar
axis: (0 veya 1, default = 0)

Sıralamanın satırlar (0) veya sütunlar (1) üzerinde yapılacağını belirler. Genellikle satırlar sıralanır.
Örnek:
df.sort_values(by="age", axis=0)  # Satırları "age" sütununa göre sıralar (standart kullanım)
df.sort_values(by="age", axis=1)  # Sütunları "age" değerine göre sıralar
ascending: (bool veya list, default = True)

Sıralamanın artan (True) veya azalan (False) olacağını belirtir. Birden fazla sütun sıralaması için liste de verebilirsiniz.
Örnek:
df.sort_values(by="age", ascending=True)  # "age" sütununu artan sıraya göre sıralar
df.sort_values(by="age", ascending=False)  # "age" sütununu azalan sıraya göre sıralar
df.sort_values(by=["age", "salary"], ascending=[True, False])  # "age" artan, "salary" azalan sıralanır
inplace: (bool, default = False)

True yaparsanız, orijinal DataFrame'i yerinde sıralar (yeni bir DataFrame döndürmez, mevcut olanı değiştirir).
Örnek:
df.sort_values(by="age", inplace=True)  # DataFrame'i kalıcı olarak sıralar
kind: (string, default = 'quicksort')

Kullanılacak sıralama algoritmasını belirtir. Seçenekler:
'quicksort': Hızlıdır, en yaygın kullanılan.
'mergesort': Kararlı bir sıralama yapar (aynı değerlere sahip olanlar orijinal sırayı korur).
'heapsort': Daha çok teorik avantajlar sağlar, pratikte daha az kullanılır.
Örnek:
df.sort_values(by="age", kind="mergesort")  # "age" sütununu kararlı bir şekilde sıralar
na_position: ('first' veya 'last', default = 'last')

NaN (eksik değerler) sıralanırken nereye konulacağını belirtir.
'first': NaN değerler en başa konur.
'last': NaN değerler en sona konur (varsayılan).
Örnek:
df.sort_values(by="age", na_position="first")  # NaN değerler sıralamanın başında yer alır
ignore_index: (bool, default = False)

Sıralandıktan sonra orijinal indekslerin korunup korunmayacağını belirtir. True yaparsanız, sıralamadan sonra indeksler sıfırlanır.
Örnek:
df.sort_values(by="age", ignore_index=True)  # Sıralamadan sonra indeksler sıfırlanır
key: (fonksiyon, default = None)

Her sütundaki değerlere sıralamadan önce bir fonksiyon uygulamak için kullanılır. 
Bu parametre Python'ın sorted() fonksiyonundaki key parametresine benzer.
Örnek:
df.sort_values(by="age", key=lambda x: x.abs())  # Negatif değerleri pozitif değerlere dönüştürerek sıralar
Örneklerle Açıklamalar
Örnek 1: Temel Kullanım

import pandas as pd

data = {'name': ['Ali', 'Veli', 'Ayşe', 'Fatma'],
        'age': [25, 32, 19, 29]}

df = pd.DataFrame(data)

# Yaş sütununa göre sıralama
df_sorted = df.sort_values(by='age')
print(df_sorted)
Çıktı:

    name  age
2  Ayşe   19
0  Ali    25
3  Fatma  29
1  Veli   32
Örnek 2: Azalan Sıralama
python
Kodu kopyala
df_sorted_desc = df.sort_values(by='age', ascending=False)
print(df_sorted_desc)
Çıktı:

    name  age
1  Veli   32
3  Fatma  29
0  Ali    25
2  Ayşe   19
Örnek 3: Birden Fazla Sütun ile Sıralama
python
Kodu kopyala
data = {'name': ['Ali', 'Veli', 'Ayşe', 'Fatma'],
        'age': [25, 32, 19, 25],
        'salary': [5000, 6000, 2000, 5000]}

df = pd.DataFrame(data)

# Yaş ve maaşa göre sıralama (önce yaş, yaşlar eşit olursa maaş)
df_sorted = df.sort_values(by=['age', 'salary'], ascending=[True, False])
print(df_sorted)
Çıktı:

    name  age  salary
2  Ayşe   19    2000
0  Ali    25    5000
3  Fatma  25    5000
1  Veli   32    6000
Bu parametrelerle sort_values() fonksiyonunu esnek ve güçlü bir şekilde kullanabilirsiniz!
```
