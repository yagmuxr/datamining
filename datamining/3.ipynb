```python
Agg values

sum: Toplam hesaplar.
mean: Ortalama hesaplar.
median: Medyanı hesaplar.
min: Minimum değeri bulur.
max: Maksimum değeri bulur.
std: Standart sapmayı hesaplar.
var: Varyansı hesaplar.
count: Eleman sayısını verir (eksik olmayan değerler).
nunique: Benzersiz değerlerin sayısını bulur.
prod: Çarpımını hesaplar.
first: İlk değeri alır.
last: Son değeri alır.
mad: Ortalama mutlak sapmayı hesaplar.
sem: Ortalama hatanın standart sapmasını hesaplar.
quantile: Belirtilen yüzdelik dilimdeki değeri bulur (örn: quantile(0.25) ile %25'lik dilim).
skew: Çarpıklığı (asimetriyi) hesaplar.
```
