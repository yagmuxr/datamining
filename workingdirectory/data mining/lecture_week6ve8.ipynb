```python
#tabakalı ve ağırlıklı rastgele örnekleme
#alt grupları içeren bir popülasyonu örneklemeye olanak sağlayan teknşktşr tabakalı örnekleme
#her kahve çeşidinden aynı oranda ya da aynı sayıda veri seçilsin -> tabakalı örnekleme kullanılır
top_counts = df_coffee['country_of_origin'].value_counts() #bu df olarak isterse ? ->
top_counts
#Tabakalı ve Ağırlıklı Rastgele Örnekleme
Tabakalı Örnekleme (Stratified Sampling):
! Popülasyon çeşitli gruplara ayrıldığında, her bir gruptan eşit oranda örnek almak için.
Nasıl Çalışır: Popülasyon, belirli özelliklere göre alt gruplara ayrılır (örneğin, ülke, yaş grubu veya ürün türü). Sonra her alt gruptan aynı oranda veya aynı sayıda örnek seçilir.
Örnek: Kahve çeşitlerinin ülkelere göre dağılımını temsil etmek istiyorsanız, her ülke grubundan belirli sayıda veri seçebilirsiniz. Böylece, örneğiniz her ülkenin kahve çeşitlerini dengeli bir şekilde içerir.
Ağırlıklı Rastgele Örnekleme (Weighted Random Sampling):

Tanım: Ağırlıklı rastgele örnekleme, belirli grupların diğerlerine göre daha fazla veya daha az temsil edilmesini sağlayarak örnekleme yapar. Her bir grubun popülasyon içindeki oranına göre ağırlık verilir.
Nasıl Çalışır: Her bir grubun popülasyondaki büyüklüğüne göre örnekleme yapılır, böylece büyük gruplardan daha fazla, küçük gruplardan daha az örnek alınır.
Örnek: Eğer belirli bir ülkeden çok fazla kahve verisi, diğer ülkelerden ise az sayıda veri varsa, her grubun popülasyondaki oranına göre veri seçimi yapılabilir.
```

```python
1. Simple Random Sampling (Basit Rastgele Örnekleme)
Bu yöntem, popülasyondan tamamen rastgele örnekler alır ve her birimin seçilme olasılığı eşittir. Tüm veri setinde her satırın seçilme şansı vardır.

python
Kodu kopyala
import pandas as pd

# Veri setinden %10 oranında rastgele örnek alma
simple_random_sample = df_coffee.sample(frac=0.1, random_state=2021)

# Sonuçları görüntüleme
print("Basit Rastgele Örnekleme:")
print(simple_random_sample.head())
Açıklama:
Bu örnekleme yöntemi, popülasyonun tamamında rastgele %10'luk bir örnek seçer.
Her satır eşit seçilme şansına sahiptir, ve random_state=2021 ile rastgelelik sabitlenmiştir.
2. Stratified Sampling (Tabakalı Örnekleme)
Bu yöntemde popülasyon, belirli gruplara veya katmanlara ayrılır ve her gruptan belirli oranda örnek alınır. Örneğin, ülkeler arasında dengeli bir dağılım sağlamak için kullanılabilir.

python
Kodu kopyala
# Veri setini ülkelere göre tabakalara ayırıp, her ülkeden %10 oranında rastgele örnek seçiyoruz
stratified_sample = df_coffee.groupby('country_of_origin').sample(frac=0.1, random_state=2021)

# Sonuçları görüntüleme
print("Tabakalı Örnekleme:")
print(stratified_sample['country_of_origin'].value_counts())
print(stratified_sample.head())
Açıklama:
Bu kod, country_of_origin sütununa göre gruplandırma yaparak her ülkeyi bir katman olarak ele alır.
Her ülkeden %10 oranında rastgele örnek alınır. Böylece tüm ülkelerden orantılı olarak veri çekilmiş olur.
Özellikle ülkelerin eşit temsil edilmesini istediğimiz durumlarda tabakalı örnekleme kullanılır.
3. Cluster Sampling (Kümeleme Örnekleme)
Kümeleme örnekleme yönteminde, popülasyon küçük gruplara (kümelere) ayrılır ve sadece bazı kümeler seçilerek tüm veriler bu kümeler üzerinden analiz edilir. Örneğin, bazı kahve çeşitlerini küme olarak seçip bu çeşitlere ait tüm veriyi analiz etmek isteyebiliriz.

python
Kodu kopyala
import random

# Kahve çeşitleri arasından rastgele 3 çeşit seçiyoruz
varieties = df_coffee['variety'].unique()
selected_varieties = random.sample(list(varieties), k=3)

# Sadece seçilen kümelere (çeşitlere) ait verileri filtreliyoruz
cluster_sample = df_coffee[df_coffee['variety'].isin(selected_varieties)]

# Sonuçları görüntüleme
print("Kümeleme Örnekleme:")
print("Seçilen çeşitler:", selected_varieties)
print(cluster_sample.head())
Açıklama:
Bu kod, variety sütunundaki benzersiz çeşitlerden rastgele 3 çeşit (küme) seçer.
Yalnızca bu seçilen çeşitlerdeki (kümelerdeki) tüm verileri alır.
Kümeleme örnekleme, genellikle maliyet ve zaman tasarrufu sağlamak için sadece belirli gruplar üzerinden çalışmak istediğimizde kullanılır.
Özet Karşılaştırma
Örnekleme Türü	Ne Zaman Kullanılır?	Avantajları	Dezavantajları
Basit Rastgele Örnekleme	Popülasyonun rastgele seçildiği durumlar	Kolay ve doğrudan	Bütün popülasyona eşit ulaşım sağlanamayabilir
Tabakalı Örnekleme	Grupların eşit temsil edilmesi gerektiğinde (örneğin ülkeler)	Her gruptan orantılı örnek alınır	Grupları tanımlamak ve eşit dağılım zor olabilir
Kümeleme Örnekleme	Tüm popülasyon yerine bazı gruplar üzerinde çalışılmak istendiğinde	Maliyet ve zaman tasarrufu sağlar	Kümeler homojen değilse temsil gücü azalabilir
Her yöntemin kendine özgü kullanım alanı vardır. Eğer popülasyonda homojen bir dağılım isteniyorsa, Tabakalı Örnekleme tercih edilir. Tüm popülasyona ulaşmanın mümkün olmadığı durumlarda Kümeleme Örnekleme kullanılır. Tüm popülasyon üzerinden rastgele bir kesit almak istiyorsanız Basit Rastgele Örnekleme uygundur.
```

```python
import pandas as pd
import numpy as np

df_error = pd.DataFrame(columns=['sample_size', 'relative_error'])

pop_mean = df_coffee['total_cup_points'].mean()  # Popülasyon ortalamasını hesaplayın

for num_of_rows in range(1, len(df_coffee) + 1):  # 1'den tüm veri boyutuna kadar döngü
    sample_mean = df_coffee.sample(n=num_of_rows)['total_cup_points'].mean()  # Örneklem ortalaması
    rel_err = np.abs(pop_mean - sample_mean) / pop_mean  # Göreli hata
    df_error.loc[len(df_error)] = [num_of_rows, rel_err]  # DataFrame'e ekle

df_error.plot(x='sample_size', y='relative_error', kind='line', title='Relative Error by Sample Size')  # Çizim
```

```python
#yaklaşık örnekleme dağılımları
#tüm olasılıkları gösterir:
import itertools as it #Python'un itertools kütüphanesinin product fonksiyonunu kullanmak için bu kütüphaneyi içeri aktarıyor. itertools.product() fonksiyonu, verilen tüm girişlerin kartezjen çarpımını (yani, tüm olasılıkları) hesaplar.
dice = {
    'die1' : [1,2,3,4,5,6],
    'die2' : [1,2,3,4,5,6],
    'die3' : [1,2,3,4,5,6],
    'die4' : [1,2,3,4,5,6],
}

result = pd.DataFrame(it.product(*dice.values()), columns=dice.keys()) 
#itertools.product() fonksiyonu, birden fazla listeyi (bu durumda zarların her birinin değer listesini) alır ve bunların tüm olasılık kombinasyonlarını oluşturur.
# *dice.values() ifadesi, dice sözlüğündeki tüm değerleri (zarların değerlerini) alır ve product fonksiyonuna parametre olarak verir. Bu, her bir zar için tüm kombinasyonları hesaplamak anlamına gelir.
# Örneğin, 4 zar olduğunda, her birinin 6 farklı sonucu olabilir. Bu durumda toplamda 
# 6 üzeri 4 = 1296 farklı kombinasyon vardır.
# it.product() fonksiyonu, her kombinasyonu bir tuple (ikili) olarak döndürür. Örneğin, [(1, 1, 1, 1), (1, 1, 1, 2), ...] gibi.
# pd.DataFrame() fonksiyonu, bu kombinasyonları bir pandas DataFrame'e dönüştürür. Bu, verilerin düzenli bir tablodaki gibi sıralanmasını sağlar.
# Bu durumda, her satır bir kombinasyonu, her sütun ise bir zarın sonucunu temsil eder
```

```python
# dört atışın ort
result['mean_roll'] = (result.die1 + result.die2 + result.die3 + result.die4) / 4

result
```

```python
result['mean_roll'] = result['mean_roll'].astype('category')
result['mean_roll'].value_counts(sort=False).plot(kind='bar')
```

Standart hata (SE), genellikle ortalamanın tahmini için kullanılır. Bu, ortalamanın standart hatası (SEM) olarak adlandırılır ve örnek ortalamasının popülasyon ortalamasına olan yakınlığını gösterir.

Standart Sapma: Bireysel veri noktalarının popülasyondaki yayılımını ölçer. Yani, popülasyon verilerinin ne kadar yayılmış olduğunu gösterir.
Standart Hata: Örneklem ortalamalarının değişkenliğini ölçer. Farklı örneklerden elde edilen ortalamaların popülasyon ortalamasından ne kadar sapabileceğini gösterir.

Standart Sapma, bireysel veriler arasındaki yayılımı ölçer.
Standart Hata, örneklem istatistiklerinin (örneğin, ortalama) popülasyon istatistiğinden ne kadar sapma gösterebileceğini ölçer.


## Standart Hata Nedir?

**Standart hata** (Standard Error, SE), bir örneklem ortalamasının popülasyon ortalamasına ne kadar yakın olduğunu gösteren bir ölçüdür. Başka bir deyişle, örneklemden elde edilen ortalamaların **popülasyon ortalamasına ne kadar yakın** olduğunu gösterir. Bu, örneklem verilerinin ne kadar tutarlı ve güvenilir olduğunu anlamamıza yardımcı olur.

Standart hata, **örneklem ortalamalarının** **standart sapması** olarak da düşünülebilir ve genellikle şu formülle hesaplanır:

$$
SE = \frac{\sigma}{\sqrt{n}}
$$

- **\(\sigma\)**: Popülasyonun standart sapması (veya örneklemdeki veriler için örneklem standart sapması).
- **\(n\)**: Örneklem büyüklüğü.

Standart hata, örneklem büyüklüğüyle ters orantılıdır. Yani, örneklem büyüklüğü arttıkça, standart hata azalır, çünkü daha büyük örneklemler daha güvenilir ortalamalar sağlar. 

**Özetle:** Standart hata, örneklemden elde edilen ortalamaların ne kadar değişken olduğunu ve popülasyon ortalamasına ne kadar yakın olduğunu ölçer.

---

## Göreli Hata Nedir?

**Göreli hata** (Relative Error), bir ölçümdeki hatanın, gerçek (doğru) değere oranıdır. Bu, hatanın büyüklüğünün doğru değere ne kadar yakın olduğunu gösteren bir oran veya yüzde değeridir. Göreli hata, genellikle **hatayı doğru değere böler** ve yüzde olarak ifade edilir.

Göreli hata şu şekilde hesaplanır:

$$
\text{Göreli Hata} = \frac{\left| \text{Gerçek Değer} - \text{Ölçülen Değer} \right|}{\left| \text{Gerçek Değer} \right|}
$$

Burada:
- **Gerçek Değer**: Doğru kabul edilen değer (örneğin, popülasyon ortalaması).
- **Ölçülen Değer**: Elde edilen örnekleme göre ölçülen değer (örneğin, örneklem ortalaması).

**Özetle:** Göreli hata, hatanın büyüklüğünü, gerçek değere göre ne kadar büyük olduğunu gösteren bir orandır. Bu oran, genellikle yüzde olarak ifade edilir.

---

## Standart Hata ve Göreli Hata Arasındaki Farklar:

| **Özellik**                  | **Standart Hata (SE)**                                    | **Göreli Hata**                                                |
|------------------------------|-----------------------------------------------------------|---------------------------------------------------------------|
| **Tanım**                    | Bir örneklem ortalamasının popülasyon ortalamasına ne kadar yakın olduğunu gösterir. | Ölçülen değerin, gerçek değere göre ne kadar uzak olduğunu gösterir. |
| **Hesaplama**                 | $$ \frac{\sigma}{\sqrt{n}} $$ (popülasyon standart sapması bölü karekök örneklem büyüklüğü) | $$ \frac{\left| \text{Gerçek Değer} - \text{Ölçülen Değer} \right|}{\left| \text{Gerçek Değer} \right|} $$ |
| **Kullanım Alanı**            | Örneklemden elde edilen sonuçların güvenilirliğini gösterir. | Hata oranını ve ölçüm doğruluğunu gösterir.                    |
| **Birim**                     | Aynı birime sahiptir (örneğin, kilogram, metre, vb.).     | Genellikle bir oran ya da yüzde (%) olarak ifade edilir.       |
| **Bağlantılı Olduğu Faktörler** | Örneklem büyüklüğü (büyüdükçe standart hata azalır).      | Gerçek ve ölçülen değer arasındaki fark.                       |

---

## Özet:

- **Standart hata**, bir örneklem ortalamasının ne kadar değişken olduğunu ve popülasyon ortalamasına ne kadar yakın olduğunu gösterir.
- **Göreli hata**, ölçülen bir değerin, gerçek değere olan oranını ifade eder ve hatanın büyüklüğünü anlamamıza yardımcı olur.

Genel olarak, standart hata örneklem büyüklüğü ve verinin dağılımıyla ilgilidir, göreli hata ise ölçümün doğruluğu ve hatasının gerçek değere oranıyla ilgilidir.
