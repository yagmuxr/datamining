```python
work_week_hours = df_stck['work_week_hrs']
mean = work_week_hours.mean()
mean
```

```python
beklenen_deger= 50
std = np.std(work_week_hours, ddof=1)
n = len(work_week_hours)
std_err = std/ np.sqrt(n)
std_err
```
