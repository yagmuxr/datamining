### 1. **Pairwise Comparisons (Çiftli Karşılaştırmalar)**

**Amaç**: ANOVA sonucunda, hangi grupların birbirinden farklı olduğunu belirlemek için yapılan ikili karşılaştırmalardır.

**Örnek Test**:
- **İş tatmini** seviyelerinin (**yüksek**, **orta**, **düşük**) **ücret seviyesi** üzerindeki etkisi.
- **Farklı eğitim seviyelerine sahip çalışanların** yıllık maaşları arasındaki farklar.

```python
import pingouin
pingouin.pairwise_ttests(data=df_stck, dv="converted_comp", between="job_sat")
```

### 2. **Bonferroni Correction**

**Amaç**: Çoklu testlerde **yanlış pozitif** oranını kontrol etmek için yapılan düzeltme yöntemidir.

**Örnek Test**:
- **İş tatminine** göre **yıllık ücret**, **mesai saati** ve **saatlik ücret** üzerinde yapılan çoklu karşılaştırmalar.

```python
import pingouin
pingouin.pairwise_ttests(data=df_stck, dv="converted_comp", between="job_sat", p_adjust="bonferroni")
```

### 3. **Kruskal-Wallis Test**

**Amaç**: Üç veya daha fazla bağımsız grubun **medyanlarının** farklı olup olmadığını test etmek için kullanılan **non-parametrik ANOVA** alternatifi.

**Örnek Test**:
- **Çalışanların** farklı **bölümlerine (IT, Pazarlama, Finans)** göre **yıllık performans** medyan farklarını test etme.

```python
import pingouin
pingouin.kruskal(data=df_stck, dv="converted_comp", between="job_sat")
```

### 4. **Mann-Whitney U Test (Wilcoxon-Mann-Whitney Test)**

**Amaç**: İki bağımsız örneklemin **medyanlarının** farklı olup olmadığını test etmek için kullanılan **non-parametrik** bir testtir.

**Örnek Test**:
- **Kadın ve erkek çalışanların** **yıllık maaşlarının** medyan farklarını test etmek.
- **İki farklı mağaza şubesindeki** **satış rakamlarının** medyan farklarını test etmek.

```python
import pingouin
pingouin.mannwhitney(data=df_stck, dv="converted_comp", between="job_sat")
```

### 5. **Wilcoxon Signed-Rank Test (İşaretli Sıralar Testi)**

**Amaç**: İki eşleştirilmiş (bağımlı) örneklemin **medyan** farklarını test etmek için kullanılan parametrik olmayan bir testtir.

**Örnek Test**:
- **Çalışanların** **önceki ve sonraki yıllık maaşları** arasındaki medyan farkını test etmek.
- **Bir ilaç tedavisinin** etkisini **öncesi ve sonrası** ölçümlerle karşılaştırmak.

```python
import pingouin
pingouin.wilcoxon(data=df_stck, dv="converted_comp", paired="job_sat")
```

### 6. **Ki-Kare Testi (Chi-Square Test)**

**Amaç**: Kategorik verilerde, iki ya da daha fazla kategorik değişkenin **bağımsız olup olmadığını** test etmek için kullanılan bir testtir. Ayrıca bir kategorik dağılımın teorik dağılımla uyumunu test etmek için de kullanılabilir.

**Örnek Test**:
- **Cinsiyet ile** **oy tercihi** arasındaki bağımsızlığı test etme.
- **Ülkelerdeki** **sosyal medya kullanım oranlarının** teorik dağılımla uyumlu olup olmadığını test etme.

```python
import pingouin
pingouin.chi2_contingency(data=df_stck, dv="converted_comp", between="job_sat")
```

### 7. **Two-Sample Proportion Test (İki Örneklem Oran Testi)**

**Amaç**: İki bağımsız grubun oranlarının eşit olup olmadığını test etmek için kullanılan testtir.

**Örnek Test**:
- **Kadın ve erkeklerin** **telefon satın alma oranları** arasındaki farkları test etmek.
- **İki farklı şehirde** yaşayanların **yeni ürün satın alma oranları** arasındaki farkı test etmek.

```python
import pingouin
pingouin.proportions_ztest(count=[10, 20], nobs=[100, 150])  # count: başarılı sonuçlar, nobs: toplam gözlem sayısı
```

### 8. **One-Sample Proportion Test (Tek Örneklem Oran Testi)**

**Amaç**: Tek bir örneklemde gözlemlenen oranın, belirli bir teorik oranla karşılaştırılmasıdır.

**Örnek Test**:
- **Bir okulda** öğrencilerin **başarı oranının** %70 olup olmadığını test etmek.
- **Bir şirketin** çalışanlarının %50'sinin **uzaktan çalışma** yaptığı oranını test etmek.

```python
import pingouin
pingouin.proportions_ztest(count=10, nobs=100, value=0.5)  # count: başarılı sonuçlar, nobs: toplam gözlem sayısı, value: test edilen oran
```
